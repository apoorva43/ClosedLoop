```python
%matplotlib inline

import numpy as np
from process import *
from parameters import *
from scipy.io import loadmat
import matplotlib.pyplot as plt
```

```python
# Loading EEG data
data = loadmat('Matlab/eeg_data.mat')
eeg_data = data['eeg_data']

# Processing without channel projection
num_trials = 1
epochs = []
for i in range(num_trials):
    eeg = eeg_data
    epoch = preproc1epoch(eeg)
    epochs.append(epoch)
```

```python
# Functions to plot signals for any channel
def plot_before(channel):
    ch = eeg_data[channel] 
    x = np.arange(1, 724639)
    y = ch
    plt.plot(x, y, label = 'before-processing')
    plt.rcParams['figure.figsize'] = [12, 8]
    plt.rcParams['figure.dpi'] = 100 
    plt.show()
    
def plot_after(channel):
    ch = epochs[0][channel] 
    x = np.arange(1, 724639)
    y = ch
    plt.plot(x, y, label = 'after-processing')
    plt.rcParams['figure.figsize'] = [12, 8]
    plt.rcParams['figure.dpi'] = 100 
    plt.show()
```

```python
import matplotlib
matplotlib.rcParams['agg.path.chunksize'] = 10000 # Better plots
```

```python
# Plotting for Fp1 channel - 1st channel
plot_before(0)
```

```python
plot_after(0)
```
