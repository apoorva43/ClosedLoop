```python
from readData import *

eeg_data = read_npz() 
```

```python
from process import *
from parameters import *
```

```python
num_trials = 1200
epochs = []
for i in range(num_trials):
    eeg = eeg_data[i]
    epoch = preproc1epoch(eeg)
    #print(epoch.shape)
    epochs.append(epoch)
```

```python
print(len(epochs))
print(epochs[0].shape)
```

```python
# Visualization

%matplotlib inline
import matplotlib.pyplot as plt
```

```python
x = np.arange(1, 551)
y = epochs[0].transpose()
plt.plot(x, y, label='epoch-1')
plt.show()
```

```python
# 1st 4 epochs

y0 = epochs[0].transpose()
y1 = epochs[1].transpose()
y2 = epochs[2].transpose()
y3 = epochs[3].transpose()
plt.plot(x, y0, label='epoch-1')
plt.plot(x, y1, label='epoch-2')
plt.plot(x, y2, label='epoch-3')
plt.plot(x, y3, label='epoch-4')
plt.legend(['epoch-1','epoch-2','epoch-3','epoch-4'])
plt.ylim([-120, 120])
plt.show()
```

```python
# average value across all 550 samples

new_epoch = []
num_samples = 550

for i in range(num_samples):
    value = 0
    for j in range(num_trials):
        value += epochs[j][0][i]
    value /= num_trials
    new_epoch.append(value)
```

```python
# Plot 
plt.plot(x, new_epoch, label='epoch-avg')
plt.show()
```
